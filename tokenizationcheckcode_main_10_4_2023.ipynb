```python
import re  
from nltk.tokenize import regexp_tokenize  
  
def tokenize_column2(df, column_name):  
    # Create an empty list to store the tokens  
    tokens = []  
    # Define the regular expression pattern to match hyphenated words  
    pattern = r'\b\w+(?:-\w+)+\b|\b\w+\b'  
    # Loop over each row in the dataframe  
    for index, row in df.iterrows():  
        # Check if the value in the column is a string  
        if isinstance(row[column_name], str):  
            # Tokenize the text in the specified column using the regular expression pattern  
            text_tokens = regexp_tokenize(row[column_name], pattern)  
            # Add the tokens to the list  
            tokens += text_tokens  
    return tokens  
```

```python
import time  
  
# Function to calculate embeddings for a list of tokens  
def calculate_embeddings(tokens_list):  
    embeddings = []  
    for token in tokens_list:  
        embedding = get_embedding(token, engine="openaitestembeddingdeployment")  
        embeddings.append(embedding)  
    return embeddings  
  
chunk_size = 10  
embeddings_list = []  
  
# Process embeddings in chunks of 10 tokens  
for i in range(0, len(df_company), chunk_size):  
    tokens_chunk = df_company['split_value'][i:i+chunk_size]  
      
    retry_count = 0  
    while retry_count < 50:  
        try:  
            embeddings_chunk = calculate_embeddings(tokens_chunk)  
            embeddings_list.extend(embeddings_chunk)  
            break  
        except Exception as e:  
            print(f"Error occurred: {e}")  
            retry_count += 1  
            time.sleep(2)  # Wait for 2 seconds before retrying  
      
    # Print progress  
    progress = min(i + chunk_size, len(df_company))  
    print(f"Processed {progress} out of {len(df_company)} token.")  
  
# Add the calculated embeddings to the dataframe  
df_company['embedding'] = embeddings_list  
  
# Display the resulting dataframe  
#print(cleaned_words2)  
```
